Click run on the cell below to import required packages

```python
import turicreate as tc
```

Cick run on the cell below in order to load the hospital data

```python
data =  tc.SFrame.read_csv('diabetic_data.csv')
```

Click run on the cell below to preview data


```python
data.explore()
```

Click run on the cell below to view plots of the data

```python
data.show()
```

Build model by changing the max_iterations and max_depth numbers and then clicking run on the cell below.  
Play around with the max_iterations and max_depth to see if you can improve the AUC


```python
train_data, test_data = data.random_split(0.9)

model = tc.random_forest_classifier.create(train_data, target='readmitted',
                                           max_iterations=100,
                                           max_depth =50, class_weights='auto')


model.evaluate(test_data)
```

```python
model.get_feature_importance().print_rows(num_rows=1000)
```
