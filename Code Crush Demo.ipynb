Click run on the cell below in order to load the hospital data.  

The data includes information about patients that have visited the hospital.  We are trying to see if we can build a model from the data to predict whether a patient will be readmitted to the hospital after being discharged.  Once a patient leaves the hospital, it is very important that we make sure that they are not readmitted within a short time period.  If they are a high risk for being readmitted we may either keep them longer or provide additional support to check up on them to make sure they are okay and are following discharge instructions such as medications.  If we can predict which patients are at risk for readmission, we can reach out to particular patients and provide extra resources to make sure that they remain healthy.

Click run on the cell below to take a look at the hospital data.





```python
data =  tc.SFrame.read_csv('diabetic_data.csv')
```

Let's now take a look at all of he data that we have loaded.  Click runon the cell below to explore our data.

```python
data.explore()
```

Now let's run a model.  We can change certain model parameters to get good results.  We want to make sure our model is as good as it can be so that we can reach out and provide much needed support to the right patients.  

Good results would mean that you are accurately predicting whether a patient will be readmitted.  One thing to think about is whether it is more important to not falsely predict a readmitted patient or whether it is more important to not miss predicting a patient that will be readmitted.

You can change your model by changing the max_iterations and max_depth numbers.  

Change the model until you get the best results you can.  

Once you have chosen your max_depth and max_iterations number, go ahead and click run on the cell below

The maximum depth is how deep the tree is.
The number of iterations is how many trees we build.




```python
train_data, test_data = data.random_split(0.9)

model = tc.random_forest_classifier.create(train_data, target='readmit',
                                           max_iterations=10,
                                           max_depth =3, class_weights='auto')


```

```python
model.evaluate(test_data)
```
